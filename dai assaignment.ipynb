```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Valuation'].str.replace('$', '').str.replace('B', '').astype(float), kde=True)
plt.title('Distribution of Company Valuation (in Billion $)')
plt.xlabel('Valuation ($B)')
plt.show()
```

```python
df['Year Joined'] = pd.to_datetime(df['Date Joined']).dt.year
plt.figure(figsize=(8, 5))
sns.countplot(x='Year Joined', data=df, palette='viridis')
plt.title('Number of Companies Joined by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(y='Industry', data=df, order=df['Industry'].value_counts().index, palette='muted')
plt.title('Number of Companies by Industry')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(y='Country', data=df, order=df['Country'].value_counts().index, palette='cool')
plt.title('Number of Companies by Country')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Continent', data=df, palette='coolwarm')
plt.title('Number of Companies by Continent')
plt.show()
```

# Bivariate Analysis


```python
# 1. Valuation vs Industry
plt.figure(figsize=(10, 6))
sns.barplot(x='Industry', y=df['Valuation'].str.replace('$', '').str.replace('B', '').astype(float), data=df, ci=None, palette='viridis')
plt.title('Average Valuation by Industry')
plt.xticks(rotation=45)
plt.ylabel('Valuation ($B)')
plt.show()
```

```python
# 2. Valuation vs Country
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y=df['Valuation'].str.replace('$', '').str.replace('B', '').astype(float), data=df, ci=None, palette='cool')
plt.title('Average Valuation by Country')
plt.xticks(rotation=45)
plt.ylabel('Valuation ($B)')
plt.show()
```

```python
# 3. Funding vs Valuation (Scatter Plot)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Funding', y='Valuation', data=df)
plt.title('Funding vs Valuation')
plt.xlabel('Funding ($B)')
plt.ylabel('Valuation ($B)')
plt.show()
```

```python
# 4. Valuation over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date Joined', y='Valuation', data=df, marker='o')
plt.title('Valuation Over Time')
plt.xlabel('Date Joined')
plt.ylabel('Valuation ($B)')
plt.show()
```
